```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import joblib

df = pd.read_csv("train_u6lujuX_CVtuZ9i.csv")

df.dropna(inplace=True)

df.drop(columns=["Loan_ID" , "Gender" ], inplace=True)






label_cols = [ 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status']
label_encoders = {}


for col in label_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le 
    


X = df.drop("Loan_Status", axis=1)
y = df["Loan_Status"]


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.35, random_state=20)


knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2, weights='uniform')
knn.fit(X_train, y_train)
knn_accuracy = accuracy_score(y_test, knn.predict(X_test))
print("KNN Accuracy:", knn_accuracy)


joblib.dump(knn, 'kn_model.pkl')

logreg = LogisticRegression(max_iter=2000, C=1.0, solver='liblinear')

logreg.fit(X_train, y_train)
logreg_accuracy = accuracy_score(y_test, logreg.predict(X_test))
print("Logistic Regression Accuracy:", logreg_accuracy)


joblib.dump(logreg, 'logreg_model.pkl')


rf = RandomForestClassifier(n_estimators=1000, max_depth=10, min_samples_split=2, min_samples_leaf=1)

rf.fit(X_train, y_train)
rf_accuracy = accuracy_score(y_test, rf.predict(X_test))
print("Random Forest Accuracy:", rf_accuracy)


joblib.dump(rf, 'rf_model.pkl')

joblib.dump(label_encoders, 'label_encoders.pkl')  
joblib.dump(scaler, 'scaler.pkl') 

print("Models and encoders have been saved successfully.")



accuracies = {
    "K-Nearest Neighbors": knn_accuracy,
    "Logistic Regression": logreg_accuracy,
    "Random Forest": rf_accuracy
}

joblib.dump(accuracies, 'model_accuracies.pkl')
print("Accuracies saved.")
```
